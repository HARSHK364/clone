```python
usa=pd.read_csv("USA_Housing.csv")
                
```

```python
usa
```

```python
usa.pop("Address")
```

```python
usa.info()
```

```python
X=usa[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y=usa[["Price"]]
```

```python
sb.displot(usa["Price"])
```

```python
usa.isnull().sum()
```

```python
sb.heatmap(usa.isnull())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.35)
```

```python
len(X_test)/5000
```

```python
usalr=LinearRegression()
```

```python
usalr.fit(X_train,y_train)
```

```python
usalr.coef_
```

```python
pd.DataFrame(usalr.coef_,index=X.columns,columns=["my value"])
```

```python
usalr.intercept_
```

```python
usalr.predict([100000,20,5,3,1233])
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
ypr=usalr.predict(X_test)
```

```python
mean_absolute_error(y_test,ypr)
```

```python
mean_squared_error(y_test,ypr)
```

```python
np.sqrt(mean_squared_error(y_test,ypr))
```

```python
#more the error less accurate model
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,ypr)
```
